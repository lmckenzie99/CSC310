This dataset contains demographic information that relates education, in years, to job level at a bank. The set also expands on this by including other demographic markers, such as gender and whether or not the person is representing a minority group.

There are 4 variables being tracked in this data set:
- Job
  - Job level of employee
- Education
  - How many years of education the employee has completed
- Gender
  - Male/Female
- Minority
  - Whether or not the employee represents a minority group

## Dataset: Bank Wages (R Datasets)

There are a total of 5 columns.  
There are a total of 474 rows.  
Columns are both categorical (3) and numerical (1).
One column has unused data, rownames.
There is one column that provides no information, for the purpose of graphing certain values it was removed.  

- The mode for each categorical column:
  |Job|Gender|Minority|
  |---|------|--------|
  |admin|male|no|

- The Mean, Max, and Min values, *in years*, for the numerical column, education:
  |   |Mean|Max|Min|
  |---|------|---|---|
  |Education|13.49|21|8|

*Histograms and bar graphs, along with synopsis can be found at the bottom of this Jupyter Notebook*


```python
import pandas as pd
url = 'https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/refs/heads/master/csv/AER/BankWages.csv'
wagesDF = pd.read_csv(url)
```

```python
wagesDF.head()
```

```python
wagesDF.shape
```

```python
wagesDF.columns
```

```python
wagesDFnew = wagesDF.drop('rownames', axis=1)
```

```python
wagesDFnew.head()
```

```python
wagesDFnew.shape
```

```python
wagesDFnew.columns
```

```python
categoricalColumns = ['job','gender','minority']
for column in categoricalColumns:
  print(f"Mode of column {column}:\n")
  print(wagesDFnew[column].mode())
  print('\n')
```

```python
for column in categoricalColumns:
  print(wagesDFnew[column].value_counts())
  print('\n')
```

```python
numericalColumn = wagesDFnew['education']
meanVal = numericalColumn.mean()
maxVal = numericalColumn.max()
minVal = numericalColumn.min()

print(f"Mean: {meanVal:.2f}\nMax: {maxVal}\nMin:{minVal}")
```
